```python
# used as plot program
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(figsize=(12,4), dpi=300,facecolor=(1, 1, 1))#figsize
# plt.tight_layout()#
plt.subplots_adjust(wspace =0.35, hspace =0.42)#


plt.subplot(121)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="iRNA-Methyl_f-score.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#
all_results
x=range(len(all_results))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotataxytext
plt.annotate("["+str(max_value_index+1)+","+str(max_value)+"]",xytext=(max_value_index-0.2*length_x,max_value),xy=(max_value_index,max_value),fontsize=11)
plt.plot([max_value_index,max_value_index],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#
plt.plot(x,all_results)#
plt.title("f_classif (Dataset-S51)",fontsize=11)#
plt.legend(loc="upper right",fontsize=11)#legend

#

plt.xlim([0, length_x]) #x
plt.ylim((0.5, 1))#y
#
plt.ylabel('Accuracy',fontsize=11) #y
plt.xlabel('Feature dimension',fontsize=11)#x
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #y
my_x_ticks = np.arange(0, length_x, (length_x/5)/10*10)
plt.xticks(my_x_ticks)#x
plt.text(-0.19*length_x*1.0,0.98,'(A)',fontsize='xx-large') #


plt.subplot(122)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="iRNA-Methyl_mrmr.txt"
all_results=[]
with open(file_location,'r') as f:#r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))
all_results
x=range(len(all_results))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)
plt.annotate("["+str(max_value_index+1)+","+str(max_value)+"]",xytext=(max_value_index-0.2*length_x,max_value),xy=(max_value_index,max_value),fontsize=11)
plt.plot([max_value_index,max_value_index],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)
plt.plot(x,all_results)
plt.title("mrmr (Dataset-S51)",fontsize=11)
plt.legend(loc="upper right",fontsize=11)


#

plt.xlim([0, length_x])
plt.ylim((0.5, 1))
#
plt.ylabel('Accuracy',fontsize=11)
plt.xlabel('Feature dimension',fontsize=11)
#
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks)
my_x_ticks = np.arange(0, length_x, (length_x/5)/10*10)
plt.xticks(my_x_ticks)
plt.text(-0.19*length_x*1.0,0.98,'(B)',fontsize='xx-large')

plt.show()#plt.savefig

# plt.savefig('2RFH-feature_selection_chart.png',bbox_inches='tight') # pdf, svg,tiff bbox_inches
# print "done"
```
