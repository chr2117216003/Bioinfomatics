```python
# example program
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab as pl
#data preprocessing
iris = sns.load_dataset("iris") #iris
data=iris[["sepal_length","sepal_width"]]# sepal_length  sepal_width
data.fillna(0.1)#  Nan Na, nan 0.1
# print data
# print data
# data=data.values
# print data
data=data[0:20]#  20
data.index=["1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3"] #index,indexy
#define a panel with size (12,12) and 300 dpi
plt.figure(figsize=(12,12), dpi=300,facecolor=(1, 1, 1))# 

#adjust distrances of subplots
# plt.subplots_adjust(wspace =0.3, hspace =0.15)

#plot multi sub figures into one figure
ax=sns.heatmap(data,linewidths=.5,cmap="YlOrRd")  #datalinewidth=.5cmap
ax.set_xlabel('')# x
plt.yticks(Rotation=0) #y
plt.ylabel("heatmap_seaborn",fontsize=16) #y
#show figure 
plt.show()# 
# plt.savefig('test_boxplot_seaborn.pdf',bbox_inches='tight') #plt.show(),
```
